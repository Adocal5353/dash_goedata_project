```python
from selenium import webdriver
from bs4 import BeautifulSoup,Comment
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Exécuter en mode headless

driver = webdriver.Chrome(options=options)

driver.get("https://perspective.usherbrooke.ca/bilan/servlet/BMTendanceStatPays?codeTheme=6&codeStat=EG.USE.ELEC.KH.PC&codePays=TGO&optionsPeriodes=Aucune&codeTheme2=1&codeStat2=SP.POP.GROW&codePays2=TGO&optionsDetPeriodes=avecNomP")

# Attendre que la page se charge complètement (optionnel)
element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//table//td"))
)

# Extraire le contenu de la balise <body>
page_source = driver.page_source
soup = BeautifulSoup(page_source,'html.parser')
driver.quit()

# Fonction pour trouver et supprimer les commentaires
def remove_comments(my_element):
    for child in my_element.children:
        if isinstance(child, Comment):
            child.extract()  # Supprimer le commentaire
        elif child.name:
            remove_comments(child)  # Récursivement vérifier les enfants

# Supprimer les commentaires du code HTML
remove_comments(soup)

# Trouver le tbody contenant les données du tableau
tbody = soup.find('tbody', id='listeDonnee')

# Trouver toutes les cellules du tableau
cells = tbody.find_all('td')

# Extraire les données du tableau
annees = []
valeurs = []
for i, td in enumerate(cells):
    if i % 3 == 0:
        annees.append(td.text)
    elif i % 3 == 1:
        valeurs.append(td.text)

# Créer un DataFrame pandas avec les données séparées en colonnes
df = pd.DataFrame({'Année': annees, 'Valeur': valeurs})


# Afficher le DataFrame
display(df)
```
