```python
import osmnx as ox
import matplotlib.pyplot as plt

# Chemin vers votre fichier OSM
chemin_fichier_osm = "fichier_osm/map.osm"

# Charger le fichier OSM
graph = ox.graph.graph_from_xml(chemin_fichier_osm)

# Personnalisation de la carte
fig, ax = ox.plot_graph(graph, edge_color='gray', edge_linewidth=1, node_color='red', node_size=20)

# Ajouter un titre et afficher la lgende
plt.title("Carte personnalise")
plt.legend(["Rues", "Nuds"])

# Afficher la carte
plt.show()
```
